```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = 'japan_birth.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to understand its structure
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set the aesthetics for the plots
sns.set_style("whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))
fig.suptitle('Japan Birth Data Trends', fontsize=16)

# Total Births over time
sns.lineplot(ax=axes[0, 0], x=data['year'], y=data['birth_total'], color='blue')
axes[0, 0].set_title('Total Births Over Time')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Total Births')

# Birth Rate per 1,000 people over time
sns.lineplot(ax=axes[0, 1], x=data['year'], y=data['birth_rate'], color='green')
axes[0, 1].set_title('Birth Rate Per 1,000 People Over Time')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Birth Rate')

# Total Fertility Rate over time
sns.lineplot(ax=axes[1, 0], x=data['year'], y=data['total_fertility_rate'], color='red')
axes[1, 0].set_title('Total Fertility Rate Over Time')
axes[1, 0].set_xlabel('Year')
axes[1, 0].set_ylabel('Total Fertility Rate')

# Birth Gender Ratio over time
sns.lineplot(ax=axes[1, 1], x=data['year'], y=data['birth_gender_ratio'], color='purple')
axes[1, 1].set_title('Birth Gender Ratio Over Time')
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('Birth Gender Ratio (Males per 100 Females)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust subplots to fit into the figure area.
plt.show()
```
