# Evaluación del Ajuste en un Molde de Inyección de Plásticos Utilizando la Prueba T Pareada

## Introducción
En la industria de plásticos, el control de calidad es un aspecto fundamental para garantizar la uniformidad y la fiabilidad de los productos. Uno de los parámetros críticos en la producción de piezas plásticas mediante el proceso de inyección es el espesor de las mismas. Cualquier variación en el espesor podría resultar en problemas de calidad, tales como debilidad estructural o incumplimiento de especificaciones.

Este artículo se enfoca en un caso práctico donde se aplica la Prueba T pareada para evaluar el efecto de un ajuste realizado en un molde de inyección con el objetivo de mejorar la consistencia en el espesor de las piezas de plástico producidas.

## Contexto
Una fábrica de plásticos observó que había una variabilidad considerable en el espesor de un tipo específico de pieza que estaban produciendo. Para resolver este problema, se realizó un ajuste en el molde de inyección. El objetivo del estudio es determinar si este ajuste ha tenido un efecto estadísticamente significativo en la reducción de la variabilidad del espesor de las piezas.

## Metodología
### Recopilación de Datos
Se seleccionaron al azar 30 piezas antes del ajuste del molde (Grupo A) y 30 piezas después del ajuste (Grupo B). Se midió el espesor de cada pieza en milímetros utilizando un micrómetro de alta precisión.

### Resultados
Supongamos que al aplicar la fórmula de la Prueba T pareada, obtuvimos un valor de T de 10.87 y un valor p de 9.49e-12.

Dado que el valor p es menor que 0.05, rechazamos la hipótesis nula, lo que sugiere que las diferencias en el espesor de las piezas antes y después del ajuste en el molde son estadísticamente significativas.

## Interpretación y Conclusión
El resultado de la Prueba T pareada indica que el ajuste realizado en el molde de inyección ha tenido un impacto estadísticamente significativo en el espesor de las piezas de plástico. En términos prácticos, esto podría significar una mejora en la calidad y la uniformidad del producto, lo que a su vez podría llevar a una reducción de costos por defectos y devoluciones, y potencialmente a un aumento en la satisfacción del cliente.

Es esencial destacar que si bien la prueba estadística demuestra que el cambio es significativo, siempre es recomendable realizar pruebas adicionales y un seguimiento continuo para asegurar que las mejoras se mantienen a lo largo del tiempo y bajo diferentes condiciones de producción.

Por lo tanto, la Prueba T pareada se presenta como una herramienta valiosa para los ingenieros de procesos y los responsables de control de calidad en la industria de plásticos, permitiéndoles tomar decisiones basadas en datos empíricos y estadísticamente sólidos.

```python
# Importar bibliotecas necesarias
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
espesores_antes = np.array([2.05, 2.10, 2.00, 2.12, 2.08, 2.01, 2.09, 2.07, 2.11, 2.15,
                            2.05, 2.03, 2.08, 2.10, 2.12, 2.14, 2.09, 2.11, 2.04, 2.06,
                            2.01, 2.00, 2.02, 2.12, 2.15, 2.13, 2.07, 2.09, 2.05, 2.08])

espesores_despues = np.array([2.00, 2.01, 1.99, 2.02, 2.01, 1.99, 2.00, 2.00, 2.01, 2.00,
                              2.01, 2.00, 1.99, 2.00, 2.02, 2.01, 2.00, 2.00, 1.98, 1.99,
                              1.99, 1.98, 1.99, 2.00, 2.00, 2.01, 2.00, 2.00, 2.01, 1.99])

# Realizar la Prueba T pareada
t_stat, p_value = stats.ttest_rel(espesores_antes, espesores_despues)

# Gráfico de dispersión (Scatter plot)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(range(len(espesores_antes)), espesores_antes, label='Antes del ajuste', color='r')
plt.scatter(range(len(espesores_despues)), espesores_despues, label='Después del ajuste', color='b')
plt.title('Gráfico de Dispersión de Espesores')
plt.xlabel('Índice de la Muestra')
plt.ylabel('Espesor (mm)')
plt.legend()

# Histograma
plt.subplot(1, 2, 2)
plt.hist(espesores_antes, alpha=0.5, label='Antes del ajuste', color='r')
plt.hist(espesores_despues, alpha=0.5, label='Después del ajuste', color='b')
plt.title('Histograma de Espesores')
plt.xlabel('Espesor (mm)')
plt.ylabel('Frecuencia')
plt.legend()

# Mostrar gráficos
plt.tight_layout()
plt.show()

# Diagrama de caja (Boxplot)
plt.figure()
plt.boxplot([espesores_antes, espesores_despues], labels=['Antes del ajuste', 'Después del ajuste'])
plt.title('Diagrama de Caja de Espesores')
plt.ylabel('Espesor (mm)')
plt.show()

# Imprimir resultados de la prueba T
print(f"Estadístico T: {t_stat}")
print(f"Valor p: {p_value}")

# Evaluar si el resultado es estadísticamente significativo
nivel_significancia = 0.05

if p_value < nivel_significancia:
    print("El ajuste en el molde ha tenido un efecto estadísticamente significativo en el espesor de las piezas.")
else:
    print("El ajuste en el molde no ha tenido un efecto estadísticamente significativo en el espesor de las piezas.")
```
